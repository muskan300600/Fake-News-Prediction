```python
import pandas as pd
import numpy as np
import re #searching words in a text or paragraph
from nltk.corpus import stopwords #words which don't add much value
from nltk.stem.porter import PorterStemmer #Stemming
from sklearn.feature_extraction.text import TfidfVectorizer #convert text to feature
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
#Printing the stopwords in English
stopwords.words('english')
```

#### Data Pre-processing

```python
#Loading the dataset
data = pd.read_csv("train.csv")
data
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
#Replacing null values with empty strings
data = data.fillna('')
```

```python
#merging title and author
data['content']=data['author']+' '+data['title']
data
```

```python
X = data.drop(['label'],axis=1)
y = data.label
```

```python
X.shape,y.shape
```

```python
X
```

Stemming : It is a process of reducing a word to its root word. 

```python
port_stem = PorterStemmer()
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
data['content']=data['content'].apply(stemming)
```

```python
X = data.content.values
y=data.label.values
```

```python
X.shape,y.shape
```

```python
X
```

```python
import nltk
nltk.download('punkt')
```

```python
# Converting texts to numbers

vectorizer = TfidfVectorizer()

vectorizer.fit(X)
X = vectorizer.transform(X)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_train)
score = accuracy_score(pred,y_train)
```

```python
score
```

```python
pred1 = lr.predict(X_test)
score1 = accuracy_score(pred1,y_test)
```
